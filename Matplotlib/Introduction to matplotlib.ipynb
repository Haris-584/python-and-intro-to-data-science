## Matplotlib example workflow

```python
# import library
%matplotlib inline
import matplotlib.pyplot as plt

# 1.prepare data 
x = [1,2,3,4]
y = [11,223,3,424]

# 2.setup plots
fig , ax = plt.subplots(figsize=(10,10))

# 3.plot data
ax.plot(x,y)

# 4.Customize plot 
ax.set(title= "Sample plot", xlabel="X-AXIS", ylabel="Y-AXIS")

# 5. Save and show 
fig.savefig("sample-plot.png")
```

## Making figures with numpy arrays

```python
import numpy as np
x = np.linspace(0,10,100)
x[:10]
```

```python
# line graph
fig ,ax = plt.subplots()
ax.plot(x , x**2)
```

```python
# scatter graph
fig , ax = plt.subplots()
ax.scatter(x , np.sin(x))
```

Make a plot from dictionary

```python

nut_dict = {"Almont" : 400, "peanut":200, "Cashew":800}
fig , ax = plt.subplots()
ax.bar(nut_dict.keys(), nut_dict.values())
ax.set(title="Dry fruit prices" , ylabel="Prices (RS)")
```

```python
fig , ax = plt.subplots()
ax.barh(list(nut_dict.keys()), list(nut_dict.values()))
ax.set(title="Dry fruit prices" , xlabel="Prices (RS)")
```

```python
import pandas as pd
heart_disease = pd.read_csv("heart-diseases.csv")
over_50 = heart_disease[heart_disease["age"] > 50]

fig, ax = plt.subplots(figsize=(10, 6))
scatter = ax.scatter(over_50["age"], 
                     over_50["chol"], 
                     c=over_50["target"], 
                     cmap="winter")
ax.set(title="Heart Disease and Cholesterol Levels",
       xlabel="Age",
       ylabel="Cholesterol")
ax.axhline(y=over_50["chol"].mean(), 
           color='r', 
           linestyle='--', 
           label="Average");
ax.legend(*scatter.legend_elements(), title="Target");
```
